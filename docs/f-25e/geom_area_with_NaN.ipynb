# Handling NaN values in geom_area()
`geom_area()` will now break if `y` includes `NaN` values.

```python
from lets_plot import *
from lets_plot.bistro import *
from scipy.stats import multivariate_normal
from lets_plot.geo_data import *
import pandas as pd
import numpy as np
```

```python
LetsPlot.setup_html()
```

```python
df_full = pd.DataFrame({
  'x': [2, 6, 8, 9, 5, 7, 4, 3, 1, 0],  
  'y': [3, 5, 6, 8, 6, 7, 4, 5, 4, 2]
})

df_x = pd.DataFrame({
  'x': [2, 6, 8, 9, None, 7, 4, 3, 1, None],  
  'y': [3, 5, 6, 8, 6, 7, 4, 5, 4, 2]
})

df_y = pd.DataFrame({
  'x': [2, 6, 8, 9, 5, 7, 4, 3, 1, 0],  
  'y': [3, 5, 6, 8, None, 7, 4, 5, 4, None]
})
```

```python
full = ggplot(df_full, aes('x', 'y')) + geom_area()
x_na = ggplot(df_x, aes('x', 'y')) + geom_area()
y_na = ggplot(df_y, aes('x', 'y')) + geom_area()
```

```python
gggrid([full, x_na, y_na], ncol=3)
```
